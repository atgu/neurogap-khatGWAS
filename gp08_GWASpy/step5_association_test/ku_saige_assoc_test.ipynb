```python
import hail as hl
import hailtop.batch as hb
```

```python
# get file size to allocate resources for batch jobs accordingly
def get_file_size(file):
    file_info = hl.utils.hadoop_stat(file)
    size_bytes = file_info['size_bytes']
    size_gigs = size_bytes / (1024 * 1024 * 1024)
    return size_gigs
```

```python
# perform single-variant association tests per chromosome
## LOCO=True means leave one chromosome out - it's highly recommended to avoid proximal contamination

# Binary trait
def association_test_binary(b, phenotype, site, plink_files, rda_file, variance_ratio, chr, storage_size):
    j = b.new_job(name=f'chr{chr} - {phenotype} - {site}')
    j.image('wzhou88/saige:1.3.0')
    j.cpu(16)
    j.storage(f'{storage_size}Gi')

    # Declare a real output resource
    j.declare_resource_group(ofile={'txt': 'saige_assoc.txt'})

    j.command(f'''
    set -ex

    step2_SPAtests.R \
        --bedFile={plink_files.bed} \
        --bimFile={plink_files.bim} \
        --famFile={plink_files.fam} \
        --AlleleOrder=alt-first \
        --GMMATmodelFile={rda_file} \
        --varianceRatioFile={variance_ratio} \
        --SAIGEOutputFile={j.ofile['txt']} \
        --LOCO=TRUE \
        --chrom {chr} \
        --is_output_moreDetails=TRUE \
        --is_Firth_beta=TRUE \
        --pCutoffforFirth=0.05
    ''')

    return j, j.ofile['txt']
```

```python
# Quantitative trait - Not actually running it, but keeping just in case
def association_test_quantitative(b, phenotype, site, plink_files, rda_file, variance_ratio, chr, storage_size):
    j = b.new_job(name=f'chr{chr} - {phenotype} - {site}')
    j.image('wzhou88/saige:1.3.0')
    j.cpu(16)
    j.storage(f'{storage_size}Gi')

    j.declare_resource_group(ofile={'txt': 'saige_assoc.txt'})

    j.command(f'''
    set -ex

    step2_SPAtests.R \
        --bedFile={plink_files.bed} \
        --bimFile={plink_files.bim} \
        --famFile={plink_files.fam} \
        --AlleleOrder=alt-first \
        --GMMATmodelFile={rda_file} \
        --varianceRatioFile={variance_ratio} \
        --SAIGEOutputFile={j.ofile['txt']} \
        --LOCO=TRUE \
        --chrom {chr} \
        --is_output_moreDetails=TRUE
    ''')

    return j, j.ofile['txt']
```

```python
PER_CHR_DIR = "gs://neurogap-bge-imputed-regional/nico/khat_gwas/gp_missingness_dataset/gp08/gwaspy/per-chr"

if __name__ == '__main__':

    backend = hb.ServiceBackend(
        billing_project='neale-pumas-bge',
        remote_tmpdir='gs://neurogap-bge-imputed-regional/nico/khat_gwas/tmp'
    )

    phenotype_cols = ["assist_khat"]
    sites = ["UCT"]
    #sites = ["AAU", "KEMRI", "Moi", "Uganda", "UCT"]

    for phenotype in phenotype_cols:
        for site in sites:

            b = hb.Batch(
                backend=backend,
                name=f'SAIGE step2 per-chr {phenotype} - {site}'
            )

            rda_file = b.read_input(
                f'gs://neurogap-bge-imputed-regional/nico/khat_gwas/saige_step1_outputs/gwaspy/{phenotype}/gp08_{site}_{phenotype}_saige_null.rda'
            )

            variance_ratio = b.read_input(
                f'gs://neurogap-bge-imputed-regional/nico/khat_gwas/saige_step1_outputs/gwaspy/{phenotype}/gp08_{site}_{phenotype}_saige_null.varianceRatio.txt'
            )

            for chr in range(1, 23):

                plink_files = b.read_input_group(
                    bed=f"{PER_CHR_DIR}/{site}_passed_all_qc_chr{chr}.bed",
                    bim=f"{PER_CHR_DIR}/{site}_passed_all_qc_chr{chr}.bim",
                    fam=f"{PER_CHR_DIR}/{site}_passed_all_qc_chr{chr}.fam",
                )

                # storage sizing: simple safe default (tune later)
                storage_size = 60

                if phenotype == "assist_khat":
                    j, out = association_test_binary(
                        b, phenotype, site, plink_files, rda_file, variance_ratio, chr, storage_size
                    )
                else:
                    j, out = association_test_quantitative(
                        b, phenotype, site, plink_files, rda_file, variance_ratio, chr, storage_size
                    )

                b.write_output(
                    out,
                    f'gs://neurogap-bge-imputed-regional/nico/khat_gwas/saige_step2_outputs/gp08/gwaspy/{phenotype}/gp08_{site}_chr{chr}_{phenotype}_saige_assoc.txt'
                )

            b.run(wait=False)
```
