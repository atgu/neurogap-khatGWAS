```python
%%bash

cat > /home/hail/fusion_data/run_kuf_fusion.sh <<'EOF'
#!/usr/bin/env bash
set -euo pipefail

IMAGE="nsvalencia/fusion-twas:v5-coloc"

HOST_FUSION_DIR="/home/hail/fusion_data"
HOST_SUMSTATS="/home/hail/kuf_with_Z_fixed_1kgp_rsID_subset.tsv"
HOST_OUT_DIR="$HOST_FUSION_DIR/out/kuf"

GCS_OUT_DIR="gs://neurogap-bge-imputed-regional/nico/khat_gwas/fusion/out/studysite_covarAll/kuf"

CONTAINER_DATA_DIR="/data"
CONTAINER_SUMSTATS_DIR="/sumstats"

LDREF_PREFIX="$CONTAINER_DATA_DIR/ref/1000gp_eur/LDREF/1000G.EUR."

MAX_JOBS=4

mkdir -p "$HOST_OUT_DIR"

cd "$HOST_FUSION_DIR"

sync_outputs_to_gcs() {
    echo "Syncing outputs to GCS..."
    gsutil -m rsync -r "$HOST_OUT_DIR" "$GCS_OUT_DIR" || \
        echo "WARNING: GCS sync failed, continuing anyway" >&2
}

dataproc_heartbeat() {
    echo "Running lightweight heartbeat..."

    if command -v hadoop >/dev/null 2>&1; then
        hadoop fs -ls / >/dev/null 2>&1 || true
    fi

    if command -v yarn >/dev/null 2>&1; then
        yarn application -list >/dev/null 2>&1 || true
    fi

    date > "$HOST_OUT_DIR/.last_heartbeat.txt" || true
}

echo "Detecting FUSION.assoc_test.R inside Docker image..."

FUSION_R=$(docker run --rm "$IMAGE" bash -lc \
'find / -name FUSION.assoc_test.R 2>/dev/null | head -n 1')

if [[ -z "${FUSION_R:-}" ]]; then
    echo "ERROR: Could not find FUSION.assoc_test.R inside $IMAGE"
    exit 1
fi

echo "Using FUSION script:"
echo "  $FUSION_R"

run_fusion_chr() {
    local host_weights_pos="$1"
    local host_out_prefix="$2"
    local chr="$3"

    local sumstats_base
    local weights_base
    local out_base
    local container_sumstats
    local container_weights
    local container_out_prefix
    local status_file

    sumstats_base=$(basename "$HOST_SUMSTATS")
    weights_base=$(basename "$host_weights_pos")
    out_base=$(basename "$host_out_prefix")

    container_sumstats="$CONTAINER_SUMSTATS_DIR/$sumstats_base"
    container_weights="$CONTAINER_DATA_DIR/$weights_base"
    container_out_prefix="$CONTAINER_DATA_DIR/out/kuf/$out_base"

    status_file="${host_out_prefix}.chr${chr}.status.txt"

    if [[ -s "$status_file" ]] && \
       grep -q "^OK$" "$status_file" && \
       [[ -s "${host_out_prefix}.chr${chr}.dat" ]]; then
        echo "  chr${chr}: already OK, skipping"
        return 0
    fi

    echo "  chr${chr}: $sumstats_base | $weights_base"

    if docker run --rm \
        -v "$HOST_FUSION_DIR:$CONTAINER_DATA_DIR" \
        -v "$(dirname "$HOST_SUMSTATS"):$CONTAINER_SUMSTATS_DIR" \
        "$IMAGE" \
        bash -lc "cd /work && \
                  Rscript '$FUSION_R' \
                    --sumstats '$container_sumstats' \
                    --weights '$container_weights' \
                    --weights_dir '$CONTAINER_DATA_DIR' \
                    --ref_ld_chr '$LDREF_PREFIX' \
                    --coloc_P 0.05 \
                    --chr '$chr' \
                    --out '${container_out_prefix}.chr${chr}.dat'"
    then
        echo "OK" > "$status_file"
        return 0
    else
        echo "FAIL" > "$status_file"
        echo "WARNING: failed chr${chr} for $sumstats_base and $weights_base" >&2
        return 0
    fi
}

if [[ ! -f "$HOST_SUMSTATS" ]]; then
    echo "ERROR: missing sumstats file: $HOST_SUMSTATS"
    exit 1
fi

sumstats_base=$(basename "$HOST_SUMSTATS")
stem="${sumstats_base%.*}"

sync_outputs_to_gcs
dataproc_heartbeat

for host_weights_pos in "$HOST_FUSION_DIR"/GTExv8.ALL.*.pos; do
    [[ -e "$host_weights_pos" ]] || continue

    wbase=$(basename "$host_weights_pos")

    [[ "$wbase" == *.nofilter.pos ]] && continue

    tissue_base="${wbase%.pos}"
    host_out_prefix="$HOST_OUT_DIR/${stem}.${tissue_base}"

    echo "============================================================"
    echo "Running KUF TWAS:"
    echo "  sumstats : $sumstats_base"
    echo "  tissue   : $wbase"
    echo "  out      : $host_out_prefix"
    echo "============================================================"

    running=0

    for chr in $(seq 1 22); do
        run_fusion_chr "$host_weights_pos" "$host_out_prefix" "$chr" &
        ((running+=1))

        if [[ "$running" -ge "$MAX_JOBS" ]]; then
            wait -n
            ((running-=1))
        fi
    done

    wait

    first=1
    rm -f "${host_out_prefix}.all_chr.dat"

    for chr in $(seq 1 22); do
        f="${host_out_prefix}.chr${chr}.dat"

        if [[ -s "$f" ]]; then
            if [[ $first -eq 1 ]]; then
                cat "$f" > "${host_out_prefix}.all_chr.dat"
                first=0
            else
                tail -n +2 "$f" >> "${host_out_prefix}.all_chr.dat"
            fi
        fi
    done

    if [[ -f "${host_out_prefix}.all_chr.dat" ]]; then
        echo "Combined output written to:"
        echo "  ${host_out_prefix}.all_chr.dat"
    else
        echo "No combined output created for:"
        echo "  ${host_out_prefix}"
    fi

    sync_outputs_to_gcs
    dataproc_heartbeat
done

sync_outputs_to_gcs
dataproc_heartbeat

echo "KUF TWAS runs finished."
EOF
```
